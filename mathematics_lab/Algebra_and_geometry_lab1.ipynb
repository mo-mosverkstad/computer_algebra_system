## Syfte

```python
import numpy as np

def gauss_elimination(A: np.ndarray, b: np.ndarray) -> np.ndarray:
    # Solve the system Ax = b using Gaussian elimination with appropriate row ordering.
    # Returns the solution x.
    # Copies of A and b so that the originals are preserved
    A = A.astype(float).copy()
    b = b.astype(float).copy()
    n: int = len(b)

    # Elimination below each pivot variable (i: row num, j: col num)
    for i in range(n):
        # To avoid division by zero or unnecessarily small values, do the following:
        # Find the row with the largest absolute value in column i among rows i,…,n-1 and move it to row i
        # First, find which row contains the largest value
        maxrow: int = i
        for j in range(i + 1, n):
            if abs(A[j, i]) > abs(A[maxrow, i]):
                maxrow = j

        # If the row is not in place i, swap both in A and b
        if maxrow != i:
            A[[i, maxrow]] = A[[maxrow, i]]
            b[[i, maxrow]] = b[[maxrow, i]]

        # Here you should add elimination of row i below each pivot variable,
        # which should eventually result in an upper triangular system/matrix.
        # Remember to update both A and b.
        for j in range(i + 1, n):
            factor: float = A[j, i] / A[i, i]
            A[j, i:] -= factor * A[i, i:]
            b[j] -= factor * b[i]

    # Here you should then add back-substitution,
    # i.e., go from the bottom and solve the unknowns step by step
    x: np.ndarray = np.zeros(n)
    for i in range(n - 1, -1, -1):
        rhs: float = b[i]
        for j in range(i+1, n, 1):
            rhs -= A[i, j] * x[j]
        x[i] = rhs / A[i, i]
    return x  # type: ignore

def matrix_dot_vector(x: np.ndarray, y: np.ndarray) -> np.ndarray:
    x_row_num = x.shape[0]
    x_col_num = x.shape[1] if len(x.shape) > 1 else 1
    y_row_num = y.shape[0]
    z: np.ndarray = np.zeros(x_row_num)
    for x_row_index in range(x_row_num):
        z[x_row_index] = sum(x[x_row_index][k] * y[k] for k in range(x_col_num))
    return z

def vector_norm(x: np.ndarray) -> float:
    norm_squared: float = 0.0
    for i in x:
        norm_squared += i ** 2
    return np.sqrt(norm_squared)

def check_solution(A: np.ndarray, x: np.ndarray, b: np.ndarray, tol: float = 1e-6) -> bool:
    # Check here if Ax ≈ b in norm within the tolerance tol (without built-in matrix functions)
    # Return True if the solution is correct within tolerance tol, otherwise False

    error: float = vector_norm(matrix_dot_vector(A, x) - b)
    print("Error:", error)
    return error <= tol
```

```python
A = np.array([[-9., -3., -7.], [-5., -2., 9.], [7., 6., -8.]], dtype=float)
b = np.array([1., 1., 1.], dtype=float)
x = gauss_elimination(A, b)
print("3x3 Solution:", x)
print("Check:", check_solution(A, x, b))

print("Check wrong solution:", check_solution(A, np.array([1., 2., 3.], dtype=float), b))
```
